```python
def optimize_SARIMA(parameters_list, d, D, s, exog):
    """
        Return dataframe with parameters, corresponding AIC and SSE

        parameters_list - list with (p, q, P, Q) tuples
        d - integration order
        D - seasonal integration order
        s - length of season
        exog - the exogenous variable
    """

    results = []

    for param in tqdm_notebook(parameters_list):
        try:
            model = SARIMAX(exog, order=(param[0], d, param[1]), seasonal_order=(param[2], D, param[3], s)).fit(disp=-1)
        except:
            continue

        aic = model.aic
        results.append([param, aic])

    result_df = pd.DataFrame(results)
    result_df.columns = ['(p,q)x(P,Q)', 'AIC']
    #Sort in ascending order, lower AIC is better
    result_df = result_df.sort_values(by='AIC', ascending=True).reset_index(drop=True)

    return result_df
```

```python
p = range(0, 4, 1)
d = 0
q = range(0, 4, 1)
P = range(0, 4, 1)
D = 0
Q = range(0, 4, 1)
s = 288
parameters = product(p, q, P, Q)
parameters_list = list(parameters)
print(len(parameters_list))
```

```python
result_df = optimize_SARIMA(parameters_list, d, D, s, train_df)
result_df
```

```python
model = SARIMAX(train_df, order=(3, 0, 1), seasonal_order=(3, 0, 1, 288)).fit()
print(model.summary())
```

# Auto ARIMA

## Train test split

```python
def split_data(df: pd.DataFrame, training: float = 0.9, validation: float = 0, test: float = 0.3):
    n = len(df)
    df_copy = df.copy()
    train_df = df_copy[0:int(n * training)]
    val_df = df_copy[int(n * training):int(n * (training + validation))]
    test_df = df_copy[int(n * (training + validation)):]
    return train_df, val_df, test_df
```

```python
train_df, val_df, test_df = split_data(df)
```

```python
plt.plot(train_df)
plt.plot(test_df)
```

```python
arima_model = auto_arima(train_df, start_p=1, start_q=1,
                         max_p=5, max_q=5,
                         d=0,
                         m=288, seasonal=True,
                         start_P=0,
                         D=0,
                         start_Q=0,
                         trace=True,
                         error_action='ignore',
                         suppress_warnings=True,
                         stepwise=True,
                         n_jobs=-1)
print(arima_model.summary())
```

```python
arima_model = auto_arima(train_df, start_p=1, start_q=1,
                         max_p=5, max_q=5,
                         d=0,
                         seasonal=False,
                         trace=True,
                         error_action='ignore',
                         suppress_warnings=True,
                         stepwise=True,
                         n_jobs=-1)
print(arima_model.summary())
```

```python
arima_model.plot_diagnostics(figsize=(15,12))
```

```python
prediction = pd.DataFrame(arima_model.predict(n_periods = 862), index=test_df.index)
prediction.columns = ['CPU usage [MHZ']
prediction
```
